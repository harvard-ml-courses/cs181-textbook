```python
import daft
```

```python
pgm = daft.PGM([5, 3])

# unobserved states
pgm.add_node(daft.Node("s1", r"$s_1$", 1, 2))
pgm.add_node(daft.Node("s2", r"$s_2$", 2, 2))
pgm.add_node(daft.Node("sdot", r"$...$", 3, 2, plot_params={'edgecolor': 'white'}))
pgm.add_node(daft.Node("sn", r"$s_n$", 4, 2))

# observed states
pgm.add_node(daft.Node("x1", r"$x_1$", 1, 1, observed=True))
pgm.add_node(daft.Node("x2", r"$x_2$", 2, 1, observed=True))
pgm.add_node(daft.Node("xdot", r"$...$", 3, 1, plot_params={'edgecolor': 'white'}))
pgm.add_node(daft.Node("xn", r"$x_n$", 4, 1, observed=True))

# state connections
pgm.add_edge("s1", "s2")
pgm.add_edge("s2", "sdot")
pgm.add_edge("sdot", "sn")
# emmision connections
pgm.add_edge("s1", "x1")
pgm.add_edge("s2", "x2")
pgm.add_edge("sn", "xn")

pgm.render(); pgm.figure.dpi = 150
```

```python
pgm.figure.savefig('../fig/HMM_DGM_asdf.png')
```

```python
pgm = daft.PGM([5, 3])

# unobserved states
pgm.add_node(daft.Node("s1", r"$s_1$", 1, 2))
pgm.add_node(daft.Node("s2", r"$s_2$", 2, 2))
pgm.add_node(daft.Node("s3", r"$s_3$", 3, 2))
pgm.add_node(daft.Node("s4", r"$s_4$", 4, 2))

# observed states
pgm.add_node(daft.Node("x1", r"$x_1$", 1, 1, observed=True))
pgm.add_node(daft.Node("x2", r"$x_2$", 2, 1, observed=True))
pgm.add_node(daft.Node("x3", r"$x_3$", 3, 1, observed=True))
pgm.add_node(daft.Node("x4", r"$x_4$", 4, 1, observed=True))

# state connections
pgm.add_edge("s1", "s2")
pgm.add_edge("s2", "s3")
pgm.add_edge("s3", "s4")
# emmision connections
pgm.add_edge("s1", "x1")
pgm.add_edge("s2", "x2")
pgm.add_edge("s3", "x3")
pgm.add_edge("s4", "x4")

pgm.render(); pgm.figure.dpi = 150
```

```python
pgm.figure.savefig('../fig/HMM_DGM_alpha_asdf.png')
```
