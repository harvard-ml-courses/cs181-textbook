```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
Xs = np.arange(-6, 6, 0.1)
Ys = [sigmoid(x) for x in Xs]
```

```python
plt.plot(Xs,Ys)
plt.savefig('../fig/log_sig_fn_GEN.png')
```

```python
Xs = np.arange(-6, 6, 0.1)
Ys = [max(0,x) for x in Xs]
```

```python
plt.plot(Xs,Ys)
plt.savefig('../fig/log_sig_fn_GEN.png')
```
