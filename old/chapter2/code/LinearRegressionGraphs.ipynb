```python
m = 1.3
b = 1
Xs = np.arange(0, 20, 1)
Ys = [(m * x) + b + (2*np.random.normal()) for x in Xs]
plt.plot(Xs,Ys, '.')
plt.savefig('../fig/data_set_scattered_GEN.jpg')
```

```python
def find_weights(X,Y):
    w = np.linalg.solve(np.dot(X.T, X) , np.dot(X.T, Y))
    return w

def make_basis(Xs, num_fns):
    X = [np.ones(Xs.shape).T]
    for i in range(1, num_fns):
        X.append(Xs**i)
    return np.array(X).T
```

```python
plt.plot(Xs,Ys, '.')
transformed_Xs = make_basis(Xs, 2)
fit_ws = find_weights(transformed_Xs, Ys)
new_Ys = np.dot(transformed_Xs, fit_ws)
plt.plot(Xs, new_Ys)
single_trans = make_basis(np.array([10]), 2)
print('single pred', np.dot(single_trans, fit_ws))
# plt.savefig('../fig/data_set_natural_fit_GEN.jpg')
```

```python
plt.plot(Xs,Ys, '.')
transformed_Xs = make_basis(Xs, 30)
fit_ws = find_weights(transformed_Xs, Ys)
new_Ys = np.dot(transformed_Xs, fit_ws)
plt.plot(Xs, new_Ys)
single_trans = make_basis(np.array([10]), 30)
print('single pred', np.dot(single_trans, fit_ws))
#plt.savefig('../fig/data_set_unnatural_fit_GEN.jpg')
```

```python
plt.plot(Xs,Ys, '.')
transformed_Xs = make_basis(Xs, 1)
fit_ws = find_weights(transformed_Xs, Ys)
new_Ys = np.dot(transformed_Xs, fit_ws)
plt.plot(Xs, new_Ys, label=r'$\lambda$ = 10000')

transformed_Xs = make_basis(Xs, 5)
fit_ws = find_weights(transformed_Xs, Ys)
new_Ys = np.dot(transformed_Xs, fit_ws)
plt.plot(Xs, new_Ys, label=r'$\lambda$ = 1')

transformed_Xs = make_basis(Xs, 30)
fit_ws = find_weights(transformed_Xs, Ys)
new_Ys = np.dot(transformed_Xs, fit_ws)
plt.plot(Xs, new_Ys, label=r'$\lambda$ = 0.00001')

plt.legend()
plt.savefig('../fig/diffregvalues_GEN.jpg')
```

```python
Xs = np.arange(0, 10, 0.1)
plt.plot(Xs, 0.0002 * np.exp(Xs), label='bias')
plt.plot(Xs, 1 / np.sqrt(Xs), label='variance')
#plt.plot(Xs, 0.5 * np.sqrt(Xs), label='variance')
plt.ylabel('Error')
plt.xlabel(r'$\lambda$')
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
plt.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=False)
plt.legend()
plt.savefig('../fig/biasvariance_GEN.jpg')
```
